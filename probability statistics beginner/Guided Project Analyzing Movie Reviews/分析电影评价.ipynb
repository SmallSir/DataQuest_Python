```python
import pandas as pd
import numpy as np
```

```python
movie = pd.read_csv('fandango_score_comparison.csv')
movie
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

## 对收视率列进行统计分析

```python
plt.hist(movie['Metacritic_norm_round'])
plt.axvline(movie['Metacritic_norm_round'].median(),c='r')
```

```python
plt.hist(movie['Fandango_Stars'])
plt.axvline(movie['Fandango_Stars'].median(),c='r')
```

### 分析：可以看出来Fandango对电影的评分大都都不低于3分，同时有接近一半的电影评分都在4分以上
### Metacritic则对电影的评分基本囊括的可以给的评分范围（1-5分)，但是主要的集中在2.5-4.5分之间

## 数据统计

```python
#Metacritic统计
Metacritic_median = movie['Metacritic_norm_round'].median()
Metacritic_mean = movie['Metacritic_norm_round'].mean()
Metacritic_std = movie['Metacritic_norm_round'].std()
print(Metacritic_median)
print(Metacritic_mean)
print(Metacritic_std)
```

```python
#Fandango统计
Fandango_median = movie['Fandango_Stars'].median()
Fandango_mean = movie['Fandango_Stars'].mean()
Fandango_std = movie['Fandango_Stars'].std()
print(Fandango_median)
print(Fandango_mean)
print(Fandango_std)
```

### 看一看出两个评论中的平均数上，由于Metacritic的低分比较多，导致其平均分和中位数都不是很高，同时方差也比较大，相比较来说Fandango的评分就比较集中，而且中位数和平均数都比较高

## 查找两个评分网站上区别最大的电影

```python
#建新列存储评分差
movie['fm_diff'] = movie['Metacritic_norm_round'] - movie['Fandango_Stars']
movie['fm_diff'] = np.abs(movie['fm_diff'])
plt.scatter(movie['Metacritic_norm_round'],movie['Fandango_Stars'])
#按照fm_diff列进行排序
movie = movie.sort_values(by = 'fm_diff',ascending=False)
movie.head()
```

## Metacritic与Fandango相关性

```python
from scipy.stats import pearsonr

r_value, p_value = pearsonr(movie["Fandango_Stars"], movie["Metacritic_norm_round"])

r_value
```

### 看一看出来两个电影评论平台的相关性差别很大

```python
#具有Metacritic_norm_roundx值和Fandango_Starsy值的线性回归
from scipy.stats import linregress

slope, intercept, r_value, p_value, stderr_slope = linregress(movie["Metacritic_norm_round"], movie["Fandango_Stars"])
```

```python
##假设有一部3分的电影在Metacritic
Fandango_number = 3.0 * slope + intercept
Fandango_number
```

## 残差图

```python
pred_1 = 1 * slope + intercept
pred_5 = 5 * slope + intercept
plt.scatter(movie["Metacritic_norm_round"], movie["Fandango_Stars"])
plt.plot([1,5],[pred_1,pred_5])
plt.xlim(1,5)
plt.show()
```
